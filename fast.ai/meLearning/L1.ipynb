```python
df_train, df_test, y_train, y_test = train_test_split(df, y, test_size=0.25)
```

```python
clf = RandomForestRegressor(n_jobs=-1)
%time clf.fit(df_train, y_train)
y_predict = clf.predict(df_test)
y_train_pred = clf.predict(df_train)
```

```python
print("Score in percentage")
print(clf.score(df_train, y_train), clf.score(df_test, y_test))
print("Log Error")
print(rmse(y_train, y_train_pred), rmse(y_test, y_predict))
```
